Data yang saya gunakan dalam tugas kali ini yaitu data mengenai pola pembelian pelanggan di pusat perbelanjaan (mall).

Berikut adalah informasi yang terkandung dalam data:

1. Nomor Faktur (invoice_no): ID unik untuk setiap transaksi yang terjadi.
2. ID Pelanggan (customer_id): ID unik untuk setiap pelanggan yang berbelanja.
3. Jenis Kelamin (gender): Jenis kelamin pelanggan, apakah laki-laki atau perempuan.
4. Usia (age): Umur pelanggan pada saat melakukan pembelian.
5. Kategori Produk (category): Jenis produk yang dibeli, seperti pakaian, sepatu, atau buku.
6. Jumlah Barang (quantity): Jumlah produk yang dibeli dalam satu transaksi.
7. Harga (price): Total biaya pembelian untuk satu transaksi.
8. Metode Pembayaran (payment_method): Metode yang digunakan untuk membayar, seperti kartu kredit, kartu debit, atau tunai.
9. Tanggal Transaksi (invoice_date): Tanggal terjadinya pembelian.
10. Pusat Perbelanjaan (shopping_mall): Nama mall tempat pelanggan melakukan pembelian.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set style
sns.set(style="whitegrid")

# 1. Bar Chart: Distribusi pembelian berdasarkan kategori produk dan gender
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_shopping_data, x='category', hue='gender', palette='pastel')
plt.title('Distribusi Pembelian Berdasarkan Kategori Produk dan Gender')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Pembelian')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()


```

```python
# 2. Bar Chart: Rata-rata pengeluaran berdasarkan usia dan kategori produk
plt.figure(figsize=(20, 10))
age_category_avg_price = customer_shopping_data.groupby(['age', 'category'])['price'].mean().reset_index()
sns.barplot(data=age_category_avg_price, x='category', y='price', hue='age', palette='coolwarm')
plt.title('Rata-rata Pengeluaran Berdasarkan Usia dan Kategori Produk')
plt.xlabel('Kategori Produk')
plt.ylabel('Rata-rata Pengeluaran (Price)')
plt.xticks(rotation=45)
plt.legend(title='Usia', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


```

```python
# 3. Bar Chart: Distribusi metode pembayaran berdasarkan kategori produk
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_shopping_data, x='category', hue='payment_method', palette='muted')
plt.title('Distribusi Metode Pembayaran Berdasarkan Kategori Produk')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Pembelian')
plt.xticks(rotation=45)
plt.legend(title='Metode Pembayaran')
plt.tight_layout()
plt.show()
```

```python
# 4. Barplot: Rata-rata pengeluaran per kelompok usia dan jenis kelamin
plt.figure(figsize=(10, 6))
sns.barplot(data=customer_shopping_data, x='age', y='price', hue='gender', ci=None, palette='muted')
plt.title('Rata-rata Pengeluaran Berdasarkan Usia dan Jenis Kelamin')
plt.xlabel('Usia')
plt.ylabel('Rata-rata Pengeluaran (Price)')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# 5. Bar Chart: Jumlah transaksi per mall
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_shopping_data, x='shopping_mall', palette='Set2')
plt.title('Jumlah Transaksi Berdasarkan Lokasi Mall')
plt.xlabel('Mall')
plt.ylabel('Jumlah Transaksi')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
